 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate synthetic data
np.random.seed(42)


A = np.random.rand(100)


B = A + np.random.normal(0, 0.1, 100)


C = 1 - A + np.random.normal(0, 0.1, 100)


D = np.random.rand(100)


df = pd.DataFrame({'A': A, 'B': B, 'C': C, 'D': D})


corr_matrix = df.corr()


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate X and Y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute Z values based on a mathematical function
z = np.sin(np.sqrt(x**2 + y**2))

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Viridis')])

# Add titles and labels
fig.update_layout(title='3D Surface Plot of sin(sqrt(x + y))',
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'
                  ))

# Show the plot
fig.show()
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Show plot in notebook (e.g., Google Colab or Jupyter)
output_notebook()

# Generate data
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Add line plot
p.line(x, y, line_width=2, color="blue", legend_label="sin(x)")

# Show the plot
show(p)
```

 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap

# Display output in notebook
output_notebook()

# Generate synthetic data
np.random.seed(42)
data = {
    'x': np.random.rand(50) * 100,
    'y': np.random.rand(50) * 100,
    'sizes': np.random.randint(10, 30, 50),
    'colors': np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], 50)
}

df = pd.DataFrame(data)
source = ColumnDataSource(df)

# Create figure
p = figure(title="Scatter Plot with Varying Sizes and Colors",
           x_axis_label='X', y_axis_label='Y', width=700, height=400)

# Add scatter plot
p.circle(x='x', y='y', size='sizes', color='colors', source=source, alpha=0.7)

# Show plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
 counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge
import pandas as pd

# Display in notebook (e.g., Jupyter or Google Colab)
output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create DataFrame and ColumnDataSource
df = pd.DataFrame({'fruits': fruits, 'counts': counts})
source = ColumnDataSource(df)

# Create figure
p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.6, source=source, color="skyblue")

# Format plot
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"

# Show plot
show(p)
```

 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

# Show plots in notebook
output_notebook()

# Given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create figure
p = figure(title="Histogram", background_fill_color="#fafafa",
           x_axis_label='Value', y_axis_label='Frequency', width=700, height=400)

# Add quad glyphs for histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Show plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.
 data_heatmap = np.random.rand(10, 10)
 x = np.linspace(0, 1, 10)
 y = np.linspace(0, 1, 10)
 xx, yy = np.meshgrid(x, y)

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.layouts import layout
from bokeh.palettes import Viridis256

output_notebook()

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten data for Bokeh's ColumnDataSource
df = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})

# Create color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=df.value.min(), high=df.value.max())

# Create figure
p = figure(title="Heatmap", x_axis_location="above",
           tools="hover,save", toolbar_location='right',
           x_range=(0,1), y_range=(0,1), width=600, height=600)

p.rect(x='x', y='y', width=0.1, height=0.1, source=df,
       fill_color=transform('value', color_mapper),
       line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0,0))
p.add_layout(color_bar, 'right')

# Show plot
show(p)
```
